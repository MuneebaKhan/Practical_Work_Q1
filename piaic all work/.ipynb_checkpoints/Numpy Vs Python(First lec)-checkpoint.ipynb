```python
#np.wheres short form if and else
salary = np.array([-1,100000,200000,300000])
```

```python
np.where(salary<0,100000,salary)
```

```python
#Jaha empty hai to wo data fill karne ke liye kuch bhi dal de(for e.g(np.meann))
np.where(salary<0,np.mean(salary),salary)
```
