```python
import pandas as pd
import numpy as np
```

```python
student_table = pd.DataFrame({
    'name': ['afzalahmed','asad','mosib','Sqib','ALi','ashfaq'],
    'age' : np.random.randint(5,25, size=[6]),
    'Per' : [70,60,50,40,30,20]
})
student_table
```

```python
def check_age(age):
    if(age < 18):
        return 'non-Adult'
    else:
        return 'Adult'
```

```python
student_table['Status'] = student_table['age'].map(check_age)
```

```python
student_table   
```

```python
def check_per(per):
    if(per < 40):
        return 'Fail'
    else:
        return 'Pass'
```

```python
student_table['Check'] = student_table['Per'].map(check_per)
```

```python
student_table
```

```python
def check_per_gra(gra):
    if(gra >= 70):
        return 'A'
    elif(gra >=50):
        return 'B'
    elif(gra >=40):
        return 'C'
    elif(gra >=30):
        return 'D'
    else:
        return 'F'
```

```python
student_table['Grade'] = student_table['Per'].map(check_per_gra)
```

```python
student_table
```

```python
def check_per_pos(pos):
    if(pos >= 70):
        return '1'
    elif(pos >= 60):
        return '2'
    elif(pos >=50):
        return '3'
    elif(pos >=40):
        return '4'
    elif(pos >=30):
        return '5'
    else:
        return '6'
```

```python
student_table['Position'] = student_table['Per'].map(check_per_pos)
```

```python
student_table
```
